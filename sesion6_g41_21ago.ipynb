```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nacimientos = pd.DataFrame({'Municipio': ['Monterrey', 'Apodaca', 'Escobedo', 'Santiago'], 'bb_nacido':[233,569,780,123], 'Moridos':[123,80,23,12]})
```

```python
nacimientos = nacimientos.sort_values('bb_nacido', ascending=0)
```

```python
plt.bar(nacimientos['Municipio'], nacimientos['bb_nacido'])
plt.title('Nacimientos en N.L en 2020')
plt.xlabel('Municipios')
plt.ylabel('Cantidad de bebes')
```

```python
personas = ['Juan', 'Pedro', 'Ana', 'Ale']
edades = [23, 18, 20, 31]
```

```python
plt.bar(personas, edades)
```

```python
# crear un lienzo donde se colocan 2 paneles y se acomodan en 2 filas y 1 columna
fig, (panel1, panel2)=plt.subplots(2,1)
fig.tight_layout() # para que tengan el mismo ancho

# grafica 1 - Nacimientos

x1 = nacimientos['Municipio']
y1 = nacimientos['bb_nacido']
panel1.bar(x1, y1)

#grafica 2 - Edades

x2 = personas
y2 = edades

panel2.bar(x2, y2)
plt.show()
```
