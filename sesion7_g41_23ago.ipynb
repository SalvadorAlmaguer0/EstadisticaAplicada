```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#pip install statsmodels

import statsmodels.formula.api as smf
```

```python
datos = pd.read_excel('modelolineal.xlsx')
```

```python
datos = datos[['Y', 'X']]
```

```python
datos
```

```python
# Grafica de dispercion
plt.scatter(datos['X'], datos['Y'])
plt.title('Grafico de dispercion de X y Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Aparentemente existe una relacion lineal entonces 
# se puede estimar minimos cuadrados ordinarios (MCO)

reg = smf.ols(formula='Y ~ X', data = datos)
results = reg.fit()
print(results.summary())
```

```python
# P-valor < alfa=5%, se rechaza H0
# Intercepto es distinto de 0
# X o Beta1, significa que en promedio cada X incrementa en I, afecta Y, y incrementa en 0.58
```
